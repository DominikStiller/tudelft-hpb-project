```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
import spiceypy as spice

from datetime import datetime
```

```python
from spice_tools import generate_lro_ephemeris, as_et, as_utc, get_lro_beta_angle
from plotting import format_plot, save_plot
```

```python
def plot_beta_angle():
    fig, ax = plt.subplots(figsize=(12, 4))

    t_start = as_et("2010 JAN 1 00:00:00")
    days = 365

    tt = np.linspace(t_start, t_start + days * 24 * 3600, 1000)
    ax.plot(pd.to_datetime([as_utc(t, sec_prec=10) for t in tt]), np.vectorize(get_lro_beta_angle)(tt))

    ax.axhline(0, c="black")

    ax.set_xlabel("Date")
    ax.set_ylabel("Sun-Beta angle []")

    plt.xticks(rotation=45, ha='right')
    ax.set_yticks(np.arange(-90, 91, 30))

    format_plot()

plot_beta_angle()
```
